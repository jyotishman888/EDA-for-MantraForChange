```python
Responses = df.Response.str.cat(sep='/')
tokens = word_tokenize(Responses)

filtered_sentence = []
stop_words = set(stopwords.words('english'))

for w in tokens:
    if w not in stop_words:
        filtered_sentence.append(w)
        
print(filtered_sentence)
```

```python
import time
item_list = filtered_sentence
new_list = []
for e in item_list:
    if e not in ('No', 'Yes','na','Nil','None','Not','nil','Nill','NIL','Ok','NO','YES','NA','na','none','Na','.','Ni','N.A','nul','Regarding','regarding'):
        new_list.append(e)
item_list = new_list
```

```python
print(new_list)
```

```python
s = new_list
listToStr = ' '.join(map(str, s))
#a, b = listToStr.split('/////')
#print(a)
#print(b)
split_up = [s.strip() for s in listToStr.split('/')]
print(split_up)
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

lem_sent = [lemmatizer.lemmatize(s) for s in split_up]
print(lem_sent)
```

```python
import nltk
from nltk.tokenize import word_tokenize


frequency_dist = nltk.FreqDist(lem_sent)
sorted(frequency_dist,key=frequency_dist.__getitem__, reverse=True)[50:300]
```
